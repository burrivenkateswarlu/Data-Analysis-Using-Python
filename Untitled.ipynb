```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
```

```python
dt = pd.read_csv("https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv")
```

```python
dt.head()
```

```python
fig,(axis1,axis2) = plt.subplots(ncols=2,figsize=(9,5))
```

```python
fig,(axis1,axis2) = plt.subplots(ncols=2,figsize=(9,5))
sns.kdeplot(dt["TV"],ax=axis1)
sns.kdeplot(dt["radio"],ax=axis1)
sns.kdeplot(dt["newspaper"],ax=axis1)
sns.kdeplot(dt["sales"],ax=axis1)
```

```python
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
```

```python
rs2 = rs.fit_transform(dt)
rs2
```

```python
fig,(axis1,axis2) = plt.subplots(ncols=2,figsize=(9,5))
sns.kdeplot(dt["TV"],ax=axis1)
sns.kdeplot(dt["radio"],ax=axis1)
sns.kdeplot(dt["newspaper"],ax=axis1)
sns.kdeplot(dt["sales"],ax=axis1)

sns.kdeplot(rs2["TV"],ax=axis2)
sns.kdeplot(rs2["radio"],ax=axis2)
sns.kdeplot(rs2["newspaper"],ax=axis2)
sns.kdeplot(rs2["sales"],ax=axis2)
```
